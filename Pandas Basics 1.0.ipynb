# Cleaning Data - Casting Datatypes and Handling Missing Values

```python

import numpy as np
#np.nan represent missing or null values in df
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
dfp=pd.DataFrame(people)
```

```python
dfp
```

```python
dfp.replace('NA',np.nan,inplace=True)
dfp.replace('Missing',np.nan,inplace=True)
```

```python
dfp.dropna()
```

```python
#here values in subset both has to be null to remove that row 
#"how" specifies how the row should be dropped
dfp.dropna(axis='index',how='all',subset=['last','email'])
```
